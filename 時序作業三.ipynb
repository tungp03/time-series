```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
y1 = np.random.normal(size = (151,1))
y2 = np.random.normal(size = (151,1))[1:151] + 0.8*np.random.normal(size = (151,1))[0:150]
y3 = np.random.normal(size = (151,1))[1:151] - 0.8*np.random.normal(size = (151,1))[0:150]
no2plt = plt.plot(y1, label='y1')
no2plt = plt.plot(y2, label='y2')
no2plt = plt.plot(y3, label='y3')
no2plt = plt.legend()
no2plt
```

```python
y2 = np.random.normal(size = (151,1))[1:151] + 0.8*np.random.normal(size = (151,1))[0:150]
y4 = 2 + np.random.normal(size = (151,1))[1:151] + 0.8*np.random.normal(size = (151,1))[0:150]
y5 = np.random.normal(size = (151,1))[1:151] + 1.25*np.random.normal(size = (151,1))[1:151]
no3plt = plt.plot(y2, label='y2')
no3plt = plt.plot(y4, label='y4')
no3plt = plt.plot(y5, label='y5')
no3plt = plt.legend()
no3plt
```

```python
Ey6 = 0/(1-0.8)
Ey7 = 0/(1+0.8)
Ey8 = 0.4/(1-0.8)
Ey9 = 0/(1-1.25)
y6 = [Ey6]
y7 = [Ey7]
y8 = [Ey8]
y9 = [Ey9]
x = Ey6
x2 = Ey7
x3 = Ey8
x4 = Ey9

for i in range(150):
    x = 0.8*x+np.random.normal(size = (151,1))[i+1][0]
    y6.append(x)
    x2 = -0.8*x2+np.random.normal(size = (151,1))[i+1][0]
    y7.append(x2)
    x3 = 0.4+0.8*x3+np.random.normal(size = (151,1))[i+1][0]
    y8.append(x3)
    x4 = 1.25*x4+np.random.normal(size = (151,1))[i+1][0]
    y9.append(x4)

no6plt = plt.plot(y6, label='y6')
no6plt = plt.plot(y7, label='y7')
no6plt = plt.plot(y8, label='y8')
no6plt = plt.legend()
no6plt
```
