```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.random.normal(size = (51,1))
b = np.random.normal(size = (51,1))
c = np.random.normal(size = (51,1))
a
```

```python
#aa = np.vstack((a, b, c))
no1plt = plt.plot(a, label='a')
no1plt = plt.plot(b, label='b')
no1plt = plt.plot(c, label='c')
no1plt = plt.legend()
no1plt
```
